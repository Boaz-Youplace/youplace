```python
#!/usr/bin/env python
import re

# 새로운 리스트
clear_titles = []
# 텍스트를 가지고 있는 리스트
for i in df['title']:
    # 영어,숫자 제거
    text = re.sub('[^0-9가-힣]',' ',i).strip() #text = re.sub('[^a-zA-Z0-9가-힣]',' ',i).strip()
    clear_titles.append(text)
print(df['title'])
print("##")
print(clear_titles)
```
